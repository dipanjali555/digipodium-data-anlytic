```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall = [17,10,10,21,26,59]
months = ['jan','fed','mar','apr','may','jan']
```

```python
plt.bar(months,rainfall,color='#1f7777b4')
```

```python
plt.style.available
```

```python
plt.style.use('ggplot') # thema
```

```python
plt.bar(months,rainfall)
```

```python
plt.pie(
    rainfall,
    labels= months,
    autopct="%1.1f%%",
    explode = (0,0,0,0,0,0.1)
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels = months,
    autopct = '%.1f%%',
    wedgeprops = {'width': .4},
    pctdistance = .8
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels = months,
    autopct = '%.1f%%',
    wedgeprops ={'width':.4,'edgecolor':'#6666ff','linewidth':2},
    pctdistance =.8
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans =True,
            meanline= True,
            meanprops={'color':'red'}
            )
plt.show()
```
